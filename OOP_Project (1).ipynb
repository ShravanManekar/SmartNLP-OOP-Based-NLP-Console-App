```python

import nlpcloud

class NLPapp:
  def __init__(self):
    self.__database={}
    self.__first_menu()

  def __first_menu(self):
    first_input=input("""
     Hi! how would you like to proceed?
    1. Not a member? Register
    2. Already a member? Login
    3. Galti se aa gaye? Exit
     """)

    if first_input=='1':
      self.__register()


    elif first_input=='2':
      self.__login()

    else:
       exit()

  def __second_menu(self):
      second_input=input("""
       Hi! how would you like to proceed?
    1. NER
    2. Language Detection
    3. Sentiment Analysis
    4. Logout
      """)
      if second_input == '1':
        self.__ner()
      elif second_input=='2':
        self.__Language_Detection()
      elif second_input=='3':
        self.__Sentiment_Analysis()

      else:
        self.__first_menu()

  def __ner(self) :
         paragraph=input('enter the paragraph:')
         search_term=input("What you want to search:")

         client = nlpcloud.Client("finetuned-gpt-neox-20b", "da878325c6b3971c830b9b04e280b9c741fab822", gpu=True, lang="en")
         response=client.entities(paragraph, searched_entity=search_term)

         print(response)
         self.__second_menu()

  def __Sentiment_Analysis(self):
         paragraph=input('enter the paragraph:')


         client = nlpcloud.Client("distilbert-base-uncased-finetuned-sst-2-english", "da878325c6b3971c830b9b04e280b9c741fab822", gpu=False)
         response=client.sentiment(paragraph)

         print(response)
         self.__second_menu()

  def __Language_Detection(self):
         paragraph=input('enter the paragraph:')


         client = nlpcloud.Client("python-langdetect", "da878325c6b3971c830b9b04e280b9c741fab822", gpu=False)
         response=client.lang_detection(paragraph)

         print(response)
         self.__second_menu()














  def __register(self):

     name=input("Enter the Name:")
     email=input("Enter the Email:")
     password=input("Enter the Password:")

     if email in self.__database:
       print("Email already exist")
     else:
      self.__database[email]=[name,password]
      print("Registeration Successfully.Now login")
      print(self.__database)
      self.__first_menu()

  def __login(self):
    email=input("Enter the registered email:")
    password=input("Enter the registered password:")

    if email in self.__database:
      if  self.__database[email][1]==password:
        print("login succesfully")
        self.__second_menu()
      else:
        print("the enter password is wrong ")
        self.__login()
    else:
      print("This is not Existed")
      self.__first_menu()



obj =NLPapp()
```
